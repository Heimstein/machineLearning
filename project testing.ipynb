```python
#import data
import csv
import numpy as np
with open('county_facts.csv', 'r') as f:
    reader = csv.reader(f)
    info = list(reader)

with open('county_results.csv', 'r') as f:
    reader = csv.reader(f)
    results = list(reader)
```
